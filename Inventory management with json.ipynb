# Inventory Management using JSON

```python
import json
import time
#updating Inventory 
fd=open('record.json','r')
js=fd.read()
fd.close()

#converts the JSON string into dictionary
record=json.loads(js)

sales=''

#displaying menu 
print('-------------------MENU---------------------')
for key in record.keys():
    print(key,record[key]['Name'],record[key]['Price'],record[key]['Quantity'])
    print('')

print('--------------------------------------------')

#taking user input 
ui_name=input("Enter your Name             :")
ui_phno=input("Enter your Number           :")
ui_pr=str(input('Enter the Prouct ID       :'))
ui_qn=int(input('Enter the Prouct Quantity :'))


if (ui_qn<=record[ui_pr]["Quantity"]):
    
#generating bill
        
    print('--------------------------------')
    print('              BILL              ')
    print("Name              :" ,record[ui_pr]['Name'])
    print("Quantity          :" ,ui_qn)
    print("Price             :" ,record[str(ui_pr)]['Price']*ui_qn)
    print('--------------------------------')
    # Applying Discount
    total_amount=record[str(ui_pr)]['Price']*ui_qn
    if total_amount > 7000:
        discount = total_amount * 0.10  # 10% discount
        print("You got a 10% discount of ₹", discount)
    elif total_amount > 5000:
        discount = 500  # Flat ₹500 discount
        print("You got a ₹500 discount!")
    
    else:
        discount = 0  # No discount applied
    
    final_amount = total_amount - discount  # Final bill after discount
    print('Final Amount      :',final_amount)
    
    #updating value
    record[ui_pr]['Quantity']=record[str(ui_pr)]['Quantity']-ui_qn
    sales= record[ui_pr]['Name']+ ","+ str(ui_qn)+','+ ui_name +','+ui_phno+","+time.ctime()+'\n'
else:
    if(record[ui_pr]['Quantity']>0):
        
        print('we only have',record[ui_pr]['Quantity'],"Quantity")
        print('would you like to purchase it')
        
    
    
        ch= str(input("Press Y/y to purchase it"))
       
        
        if(ch=='Y' or ch=='y'):
            
        #generating bill
        
            print('--------------------------------')
            print('              BILL              ')
            print("Name      :" ,record[ui_pr]['Name'])
            print("Quantity  :" ,record[ui_pr]['Quantity'])
            print("Price     :" ,record[str(ui_pr)]['Price']*record[ui_pr]['Quantity'])
            print('--------------------------------')
            # Applying Discount
            total_amount=record[str(ui_pr)]['Price']*ui_qn
            if total_amount > 7000:
                discount = total_amount * 0.10  # 10% discount
                print("You got a 10% discount of ₹", discount)
            elif total_amount > 5000:
                discount = 500  # Flat ₹500 discount
                print("You got a ₹500 discount!")
            
            else:
                discount = 0  # No discount applied
            
            final_amount = total_amount - discount  # Final bill after discount
            print('Final Amount      :',final_amount)
            
        
        #updating value
            record[ui_pr]['Quantity']=0
            sales= record[ui_pr]['Name']+ ","+ str(record[ui_pr]['Quantity'])+","+ ui_name +","+ ui_phno + "\n"
        else:
            print('thanks!')
    else :
        print("Item out of stock")
        

   

            

#updating invntory 
js=json.dumps(record)

fd=open('record.json','w')
fd.write(js)
fd.close()


#updating the sales in sales.txt file
fd=open('Sales.txt','a')
fd.write(sales)
fd.close()




```
